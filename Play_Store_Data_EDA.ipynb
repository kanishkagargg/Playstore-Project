## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
Play_Store_Data.drop_duplicates()                                                      # droping duplicate rows
````

````python
User_Reviews.drop_duplicates()                                                        # droping duplicate rows
````

````python
Play_Store_Data['Category'].unique()                                                 # checking unique values of Sentiment to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data[Play_Store_Data['Category'] == '1.9']                                # in row 10472, after analysing each column we can see that category is 1.9 which does not makes sense. And also that each and every column value must shift to its right (Then every value like size, tyoe, rating will make sense).
````

````python
Play_Store_Data.loc[10472] = Play_Store_Data.loc[10472].shift()                      # all values shifted to the right. so we need to assign value of category (which is now must be app name) to app
````

````python
Play_Store_Data['App'].loc[10472] = Play_Store_Data['Category'].loc[10472]           # assigning Category name to App to make it correct
Play_Store_Data['Category'].loc[10472] = np.nan                                      # assigning category name to a NAa value
Play_Store_Data.loc[10472]
````

````python
Play_Store_Data['Last Updated'] = pd.to_datetime(Play_Store_Data['Last Updated'])       # converting data type of to date time
Play_Store_Data['Last Updated']
````

````python
# filling NA values to appropirate values
Play_Store_Data['Type'] = Play_Store_Data['Type'].fillna('Unknown')
Play_Store_Data['Content Rating'] = Play_Store_Data['Content Rating'].fillna('Unknown')
Play_Store_Data['Current Ver'] = Play_Store_Data['Current Ver'].fillna('Z')
Play_Store_Data['Android Ver'] = Play_Store_Data['Android Ver'].fillna('Z')

User_Reviews['Sentiment_Polarity'] = User_Reviews['Sentiment_Polarity'].fillna(value= User_Reviews['Sentiment_Polarity'].mean())
User_Reviews['Sentiment_Subjectivity'] = User_Reviews['Sentiment_Subjectivity'].fillna(value= User_Reviews['Sentiment_Subjectivity'].mean())
````

````python
Play_Store_Data['Rating'].unique()                                                      # checking unique values of Rating to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data['Rating'] = Play_Store_Data['Rating'].astype(float)                     # coverting data type to float
````

````python
Play_Store_Data['Reviews'].unique()                                                     # checking unique values of Reviews to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data['Reviews'] = Play_Store_Data['Reviews'].str.replace('3.0M','3000000')   # replacing a string to an integer
Play_Store_Data['Reviews'] = Play_Store_Data['Reviews'].astype(float)                   # coverting data type to float
````

````python
Play_Store_Data['Price'].unique                                                         # checking unique values of Price to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data['Price'] = Play_Store_Data['Price'].str.replace('Everyone','0')         # replacing string to an integer
Play_Store_Data['Price'] = Play_Store_Data['Price'].str.replace('$', '')                # removing '$' from all the values to help in making into float values
Play_Store_Data['Price'] = Play_Store_Data['Price'].astype(float)                       # coverting data type to float
````

````python
User_Reviews['Sentiment'].unique()                                                   # checking unique values of Sentiment to understand its variable and finding outliers or irrelavant data
````

````python
# coverting data type to float
User_Reviews['Sentiment_Polarity'] = User_Reviews['Sentiment_Polarity'].astype(float)
User_Reviews['Sentiment_Subjectivity'] = User_Reviews['Sentiment_Subjectivity'].astype(float)
````

````python
Play_Store_Data['Size'].unique()                                                        # checking unique values of Size to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data['Size'] = Play_Store_Data['Size'].str.replace('Varies with device','0')
Play_Store_Data['Size'] = Play_Store_Data['Size'].str.replace("M","000")                # Here all MB size is coverted into a uniform kilobyte
Play_Store_Data['Size'] = Play_Store_Data['Size'].str.replace("k","")                   # removing k to make it help in converting into float
Play_Store_Data['Size'] = Play_Store_Data['Size'].replace("1,000+",1000)                # removing '+' sign
Play_Store_Data['Size'] = Play_Store_Data['Size'].astype(float)                         # coverting data type to float
Play_Store_Data['Size']
````

````python
Play_Store_Data['Installs'].unique()                                                    # checking unique values of Installs to understand its variable and finding outliers or irrelavant data
````

````python
Play_Store_Data['Installs'] = Play_Store_Data['Installs'].str.replace(',','')            # removing ',' from numbers to help it make it integer
Play_Store_Data['Installs'] = Play_Store_Data['Installs'].str.replace("+","")            # I will use the minimum of each ignoring the + sign (to know in which category they fall)
Play_Store_Data['Installs'] = Play_Store_Data['Installs'].str.replace("Free","0")        # replacing 'Free' which is string into an integer '0'
Play_Store_Data['Installs'] = Play_Store_Data['Installs'].astype(float)                  # coverting data type to float
Play_Store_Data['Installs']
````

````python
Play_Store_Data.to_csv('PlayStoreCleaned.csv',
                  index = None)
````

````python
User_Reviews.to_csv('User_ReviewsCleaned.csv',
                  index = None)
````
